```python
import pandas as pd
```

```python
df = pd.read_csv('raw_scrape.csv')
```

```python
df.veto_power.value_counts()
```

```python
df.veto_power = df.veto_power.str.strip()
df['veto_china'] = df.veto_power.str.extract(r"(China)")
df['veto_france'] = df.veto_power.str.extract(r"(France)")
df['veto_russia'] = df.veto_power.str.extract(r"(Russian Federation)")
df['veto_ussr'] = df.veto_power.str.extract(r"(USSR)")
df['veto_uk'] = df.veto_power.str.extract(r"(UK)")
df['veto_usa'] = df.veto_power.str.extract(r"(USA)")
```

```python
df.veto_power = df.veto_power.str.replace("FranceUK", "France UK")
df.veto_power = df.veto_power.str.replace("FranceUK", "France UK")
df.veto_power = df.veto_power.str.replace("ChinaRussian", "China Russian")
df.veto_power = df.veto_power.str.replace("Russian Federation", "Russian_Federation")
```
